**Task 1:**
1. Using any finance API (we use Yahoo Finance and Tiingo during sessions) download historical stock prices for Google (symbol ‘GOOG’), for the period 1 Jan 2020 to 1 Jan 2021. 
2. Save these prices to csv file with Pandas
3. Read this csv file to display data saved. 
4. Plot a simple diagram to display closing prices for GOOG stock. 

Solution: 

**Here is the plan of my work:**
1. Get token from tiingo.com
2. Connect to Tiingo API using own token
3. Get stock prices of GOOG for the period of 2020-01-01 to 2021-01-01
4. Create dataframe df with all the data retrieved from API
5. Save the retrieved data to csv file
6. Read the stock data from the saved csv file
7. Plot the graph of Google stock prices for the given period

The code for each step is given below

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
```

1. Connect to Tiingo API using own token
2. Get stock prices of GOOG for the period of 2020-01-01 to 2021-01-01

```python
url = 'https://api.tiingo.com/tiingo/daily/GOOG/prices?startDate=2020-1-1&endDate=2021-1-1'
headers = {
    'Content-Type': 'application/json',
    'Authorization' : 'Token ec3c3bbddbe56adbc978a54f8a56a11159699a48'
}

request_response = requests.get(url, headers=headers)
stock_price_data = request_response.json()

stock_price_data
```

3. create dataframe df with all the data retrieved from API

```python
df = pd.DataFrame(stock_price_data)
df
```

4. Save the retrieved data to csv file

```python
df.to_csv('Stock_prices_GOOG_2020_2021.csv')
```

5. Read the stock data from the saved csv file

```python
data_from_csv = pd.read_csv('Stock_prices_GOOG_2020_2021.csv', index_col='date')
data_from_csv
```

6. Plot the graph of google stock prices for the given period

```python
data_from_csv['close'].plot(grid=True)
plt.xlabel('Day number')
plt.ylabel('Closing price')
plt.title('GOOG price 2020-2021')
plt.show()
```
